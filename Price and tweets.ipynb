```python
import pandas as pd
import numpy as np
import OpenBlender
import json
token = '621043f29516296a6426fcf1IIuZ0VkozNBH2C9iqZtCgcraRMs7j2'

action = 'API_getObservationsFromDataset'
# ANCHOR: 'BTC to USD'
  
parameters = { 
    'token' : token,
    'id_dataset' : '5d4c3af79516290b01c83f51',
    
}
df = pd.read_json(json.dumps(OpenBlender.call(action, parameters)['sample']), convert_dates=False, convert_axes=False).sort_values('timestamp', ascending=False)
df.reset_index(drop=True, inplace=True)
df['date'] = [OpenBlender.unixToDate(ts, timezone = 'GMT') for ts in df.timestamp]
df = df.drop('timestamp', axis = 1)       
```

```python
# dataset with the prices but problem is some of them at the beginning has a different time

df
```

```python
# calculation percentage diff between close and open

df['log_diff'] = np.log(df['close']) - np.log(df['open'])
df
```

```python
df["date"].head(10)
```

```python
# adding the column target if change > 0 then 1 else 0


df['target'] = [1 if log_diff > 0 else 0 for log_diff in df['log_diff']]
df
```

```python
#put the timestamp according to the dates


format = '%d-%m-%Y %H:%M:%S'
timezone = 'GMT'
df['timestamp'] = OpenBlender.dateToUnix(df['date'], 
                                           date_format = format, 
                                           timezone = timezone)
update = df[['date', 'timestamp', 'close', 'target']]
update.head()
```

```python
# finding datasets with similar intersection

search_keyword = 'bitcoin'
df = df.sort_values('timestamp').reset_index(drop = True)
print('From : ' + OpenBlender.unixToDate(min(df.timestamp)))
print('Until: ' + OpenBlender.unixToDate(max(df.timestamp)))
OpenBlender.searchTimeBlends(token,
                             df.timestamp,
                             search_keyword)

# decided to go with another dataset with a tweets
```
